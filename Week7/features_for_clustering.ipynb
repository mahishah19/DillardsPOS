```python
import pandas as pd
import numpy as np
```

```python
# File names
feature_region = "feature_region.csv"
feature_return_rate = "feature_returnrate.csv"
feature_size = "feature_size.csv"
feature_department = "department_feature.pkl"
feature_season = "discounted_season_feature.pkl"
feature_discounted = "discounted.pkl"
```

```python
df_region = pd.read_csv(feature_region).drop("Unnamed: 0", axis=1)
df_region
```

```python
df_returnrate = pd.read_csv(feature_return_rate)
df_returnrate ## if sku not in this df, this item has not been bought before
```

```python
df_size = pd.read_csv(feature_size).drop("Unnamed: 0", axis=1)
df_size
```

```python
df_dep=pd.read_pickle(feature_department)
df_dep.columns = ["SKU","dep_1","dep_2","dep_3"]
df_dep
```

```python
df_sea=pd.read_pickle(feature_season)
df_sea
```

```python
df_dis=pd.read_pickle(feature_discounted)
df_dis
```

```python
df = df_returnrate.copy()
df = pd.merge(df, df_size)
df = pd.merge(df,df_dep)
df = pd.merge(df,df_sea)
df = pd.merge(df,df_dis[["SKU","discounted"]])
df
```

```python
df.to_csv("5_features.csv")
```
