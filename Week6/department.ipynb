```python
deptinfo = pd.read_sql("select * from group_9.deptinfo", connection)
deptinfo
```

```python
class_1_keywords = [
    "LAUREN", "TAYLOR", "LACOSTE", "ESPRIT", "DKNY", "GOTTEX", "COLEHAAN", "CARTIER", "MOSCHINO", 
    "ST JOHN", "C KLEIN", "REEBOK", "ANNASUI", "SPERRY", "CLINIQUE", "POL", "CART", "BORA", "COLOR", "ENVIRON"
]

# Class 3 Keywords (Personal/Individual Departments)
class_3_keywords = [
    "CLINIQUE", "CARTIER", "AUSTIN", "CITIZENS", "ENVIRON", "CART", "SIGRID", "LOUIS", "ETERNIT", "ECHO", "COP", "INVEST"
]

def classify_department(name):
    name = name.upper()
    if any(keyword in name for keyword in class_1_keywords):
        return 1  # Class 1: Fashion and Apparel
    elif any(keyword in name for keyword in class_3_keywords):
        return 3  # Class 3: Personal/Individual Departments
    else:
        return 2  # Class 2: Uncertain/Generic

deptinfo['Class'] = deptinfo['deptdesc'].apply(classify_department)
```

```python
deptinfo = deptinfo.rename(columns={'dept': 'DEPT'})
```

```python
deptinfo.head()
```

```python
merged_df = pd.merge(skuinfo, deptinfo, on='DEPT', how = 'outer')
merged_df = merged_df.sort_values('SKU')[['SKU','DEPT', 'Class']].reset_index()
```

```python
merged_df
```

```python
dept_class_encoded = pd.get_dummies(merged_df['Class'], columns=['Class'])
```

```python
combined_df = pd.concat([merged_df['SKU'], dept_class_encoded], axis=1)
```

```python
pickle_filename = "department_feature.pkl"
combined_df.to_pickle(pickle_filename)
```
