```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Important Notes: This file needs to use clustering_result3.csv file, which is the output file from clustering analysis and trnsact_clean.csv file, which is the output file from trnsact data cleaning.

```python
# Global Variables
trnsact_file_name = 'trnsact_clean.csv'
clusters3_file = 'clustering_result3.csv'
```

```python
TRNSACT = pd.read_csv(trnsact_file_name, sep=',')
```

```python
clusters3 = pd.read_csv(clusters3_file)
```

```python
TRNSACT.head()
```

```python
merged = pd.merge(TRNSACT, clusters3[['SKU','cluster']], on='SKU', how='inner')
merged.head()
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Assuming 'df' is your pandas DataFrame with 'ordernumber' and 'itemnumber' columns
# Transforming the DataFrame to have one transaction per row
cluster1_trnsact = merged.loc[merged['cluster'] == 0,['TRANNUM','SKU']].groupby('TRANNUM')['SKU'].apply(list).tolist()

# Encoding the transactions
te = TransactionEncoder()
te_ary = te.fit(cluster1_trnsact).transform(cluster1_trnsact)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Running Apriori algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.08, use_colnames=True)

# Generating association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Displaying the rules
rules
```
