```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# file name
SKSTINFO_file = 'skstinfo.csv'
```

```python
# read in SKSTINFO table
SKSTINFO = pd.read_csv(SKSTINFO_file, sep=',', header= None,
                       dtype={'STORE':'int', 'SKU':'int', 'COST':'float', 'RETAIL':'float', 'Unknown':'int'},
                       names=['STORE', 'SKU', 'COST', 'RETAIL', 'Unknown'])
```

```python
# show non-null value counts for all columns
SKSTINFO.info(verbose=True, show_counts=True)
# there are no null values in all columns
```

```python
# show statistics for all columns
SKSTINFO.describe()
```

```python
# drop Unknown column
SKSTINFO.drop(columns=['Unknown'], axis=1, inplace=True)
```

```python
(SKSTINFO.COST < 0.0005).sum()
# there are 351318 rows that have cost as 0
```

```python
# show the rows
SKSTINFO[SKSTINFO.COST < 0.00000005]
```

```python
(SKSTINFO.RETAIL < 0.0005).sum()
# there are 355424 rows that have cost as 0
```

```python
# show the rows
SKSTINFO[SKSTINFO.RETAIL < 0.00000005]
```

```python
SKSTINFO[SKSTINFO.RETAIL < SKSTINFO.COST]
# there are 7540289 rows that retail price is less than the cost
```

```python
SKSTINFO[SKSTINFO.RETAIL == SKSTINFO.COST]
# there are 1110913 rows that retail price is equal to the cost
```

```python
SKSTINFO[SKSTINFO.RETAIL > SKSTINFO.COST]
# there are 30578944 rows that retail price is larger than the cost
```

```python
SKSTINFO[(SKSTINFO.COST < 0.00000005) & (SKSTINFO.RETAIL < 0.00000005)]
# there are 350340 rows that both retail price and the cost have values of 0
```
